```python
import pandas as pd
import numpy as np
```

```python
#Mugdha 
# import housing file
US_housing_file = "housing price region annually.csv"
US_housing_data = pd.read_csv(US_housing_file)
US_housing = US_housing_data.rename(columns={"DATE":"Date","ASPNE":"NorthEast HP","ASPMW": "MidWest HP","ASPW":"West HP","ASPS": "South HP"})
US_housing.head()
```

```python
#US_housing["West HP"].head()
```

```python
# import income file
familyincome_file_region = "Family income by Region.csv"
income_data_region = pd.read_csv(familyincome_file_region)
income_data_region = income_data_region.rename(columns={"DATE":"Date","MAFAINUSWEA672N":"West Income", "MAFAINUSNEA672N": "NorthEast Income", "MAFAINUSSOA672N": "South Income","MAFAINUSMWA672N": "MidWest Income"})
income_data_region.head()
```

```python
housing_v_income_region = pd.merge(income_data_region, US_housing, on="Date")
housing_v_income_region.head()
```

```python
housing_v_income_region["Avg West HP Chg"] = housing_v_income_region["West HP"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
housing_v_income_region["West Income Chg"] = housing_v_income_region["West Income"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
mean_HP = housing_v_income_region["Avg West HP Chg"].mean()
mean_INC = housing_v_income_region["West Income Chg"].mean()
housing_v_income_region["Avg West HP Chg (Norm)"] = housing_v_income_region["Avg West HP Chg"] - mean_HP
housing_v_income_region["West Income Chg (Norm)"] = housing_v_income_region["West Income Chg"] - mean_INC

housing_v_income_region["Avg NorthEast HP Chg"] = housing_v_income_region["NorthEast HP"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
housing_v_income_region["NorthEast Income Chg"] = housing_v_income_region["NorthEast Income"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
mean_HP = housing_v_income_region["Avg NorthEast HP Chg"].mean()
mean_INC = housing_v_income_region["NorthEast Income Chg"].mean()
housing_v_income_region["Avg NorthEast HP Chg (Norm)"] = housing_v_income_region["Avg NorthEast HP Chg"] - mean_HP
housing_v_income_region["NorthEast Income Chg (Norm)"] = housing_v_income_region["NorthEast Income Chg"] - mean_INC

housing_v_income_region["Avg MidWest HP Chg"] = housing_v_income_region["MidWest HP"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
housing_v_income_region["MidWest Income Chg"] = housing_v_income_region["MidWest Income"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
mean_HP = housing_v_income_region["Avg MidWest HP Chg"].mean()
mean_INC = housing_v_income_region["MidWest Income Chg"].mean()
housing_v_income_region["Avg MidWest HP Chg (Norm)"] = housing_v_income_region["Avg MidWest HP Chg"] - mean_HP
housing_v_income_region["MidWest Income Chg (Norm)"] = housing_v_income_region["MidWest Income Chg"] - mean_INC

housing_v_income_region["Avg South HP Chg"] = housing_v_income_region["South HP"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
housing_v_income_region["South Income Chg"] = housing_v_income_region["South Income"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
mean_HP = housing_v_income_region["Avg South HP Chg"].mean()
mean_INC = housing_v_income_region["South Income Chg"].mean()
housing_v_income_region["Avg South HP Chg (Norm)"] = housing_v_income_region["Avg South HP Chg"] - mean_HP
housing_v_income_region["South Income Chg (Norm)"] = housing_v_income_region["South Income Chg"] - mean_INC
housing_v_income_region.head()
```

```python
housing_v_income_region=housing_v_income_region.dropna(how='any')
housing_v_income_region.head()
```

```python
housing_v_income_region.to_csv("housingincome_region.csv",  encoding="utf-8", index=False, header=True)
```

```python
US_annual_housing = "AnnualHPUS.csv"
US_annual_housing_data = pd.read_csv(US_annual_housing)
US_annual_housing_data = US_annual_housing_data.rename(columns={"DATE":"Date","MSPUS":"Avg HP"})
US_annual_housing_data.head()
```

```python
#Kate

# import US housing file
US_housing_file = "Housing Prices for US.csv"
US_housing_data = pd.read_csv(US_housing_file)
US_housing = US_housing_data.rename(columns={"MSPUS":"Avg Home Price","DATE":"Date (Quarterly)"})
US_housing.head()
```

```python
# import region housing file
housing_file_region = "Housing Prices by region.csv"
housing_data_region = pd.read_csv(housing_file_region)
housing_region = housing_data_region.rename(columns={"DATE":"Date (Quarterly)", "ASPNE": "NorthEast AHP", "ASPMW": "MidWest AHP", "ASPW": "West AHP", "ASPS": "South AHP"})
housing_region.head()
```

```python
housing_region["NorthEast AHP Chg"] = housing_region["NorthEast AHP"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
housing_region["MidWest AHP Chg"] = housing_region["MidWest AHP"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
housing_region["West AHP Chg"] = housing_region["West AHP"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
housing_region["South AHP Chg"] = housing_region["South AHP"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
mean_NE_AHP = housing_region["NorthEast AHP Chg"].mean()
mean_MW_AHP = housing_region["MidWest AHP Chg"].mean()
mean_W_AHP = housing_region["West AHP Chg"].mean()
mean_S_AHP = housing_region["South AHP Chg"].mean()
housing_region["NorthEast AHP Chg (Norm)"] = housing_region["NorthEast AHP Chg"] - mean_NE_AHP
housing_region["MidWest AHP Chg (Norm)"] = housing_region["MidWest AHP Chg"] - mean_MW_AHP
housing_region["West AHP Chg (Norm)"] = housing_region["West AHP Chg"] - mean_W_AHP
housing_region["South AHP Chg (Norm)"] = housing_region["South AHP Chg"] - mean_S_AHP

housing_region.head()
```

```python
# import US unemployment file
US_unemployment_file = "Unemployment for US.csv"
US_unemployment_data = pd.read_csv(US_unemployment_file)
US_unemployment = US_unemployment_data.rename(columns={"LRUN64TTUSQ156S":"Unemployment Rate (%)","DATE":"Date (Quarterly)"})
US_unemployment.head()
```

```python
# import region unemployment file
unemployment_file_region = "Unemployment by Region.csv"
unemployment_data_region = pd.read_csv(unemployment_file_region)
unemployment_region = unemployment_data_region.rename(columns={"LRUN64TTUSQ156S":"Unemployment Rate (%)","DATE":"Date (Quarterly)", "CNERUR": "NorthEast UE%", "CMWRUR": "MidWest UE%", "CSOUUR": "South UE%", "CWSTUR": "West UE%"})
unemployment_region.head()
```

```python
unemployment_region["NorthEast UE Chg"] = unemployment_region["NorthEast UE%"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
unemployment_region["MidWest UE Chg"] = unemployment_region["MidWest UE%"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
unemployment_region["West UE Chg"] = unemployment_region["West UE%"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
unemployment_region["South UE Chg"] = unemployment_region["South UE%"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
mean_NE_UE = unemployment_region["NorthEast UE Chg"].mean()
mean_MW_UE = unemployment_region["MidWest UE Chg"].mean()
mean_W_UE = unemployment_region["West UE Chg"].mean()
mean_S_UE = unemployment_region["South UE Chg"].mean()
unemployment_region["NorthEast UE Chg (Norm)"] = unemployment_region["NorthEast UE Chg"] - mean_NE_UE
unemployment_region["MidWest UE Chg (Norm)"] = unemployment_region["MidWest UE Chg"] - mean_MW_UE
unemployment_region["West UE Chg (Norm)"] = unemployment_region["West UE Chg"] - mean_W_UE
unemployment_region["South UE Chg (Norm)"] = unemployment_region["South UE Chg"] - mean_S_UE

unemployment_region.head()
```

```python
US_housing_v_unemployment = pd.merge(US_housing, US_unemployment, on="Date (Quarterly)")
US_housing_v_unemployment.head()
```

```python
housing_v_unemployment_region = pd.merge(housing_region, unemployment_region, on="Date (Quarterly)")
housing_v_unemployment_region.head()
```

```python
US_housing_v_unemployment["Avg Home Price Chg"] = US_housing_v_unemployment["Avg Home Price"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
US_housing_v_unemployment["Unemployment Rate Chg"] = US_housing_v_unemployment["Unemployment Rate (%)"].pct_change(periods=1, fill_method='pad', limit=None, freq=None)
mean_AHP = US_housing_v_unemployment["Avg Home Price Chg"].mean()
mean_UER = US_housing_v_unemployment["Unemployment Rate Chg"].mean()
US_housing_v_unemployment["Avg Home Price Chg (Norm)"] = US_housing_v_unemployment["Avg Home Price Chg"] - mean_AHP
US_housing_v_unemployment["Unemployment Rate Chg (Norm)"] = US_housing_v_unemployment["Unemployment Rate Chg"] - mean_UER

US_housing_v_unemployment.head()
```

```python
US_housing_v_unemployment = US_housing_v_unemployment.iloc[1:194,:]
US_housing_v_unemployment.to_excel("Housing v Unemployment for US.xlsx")
US_housing_v_unemployment.head()
```
